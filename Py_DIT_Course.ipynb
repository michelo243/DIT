```python
import matplotlib.pyplot as plt  # type:ignore


class Polynome:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    # methode privée
    def __discriminant(self):
        return (self.b**2 - (4 * self.a * self.c)) / 2 * self.a

    # méthode privée permettant de calculer y en fonction de x
    def __calculate_y(self, x):
        return self.a * x**2 + self.b * x + self.c

    def trace(self, start, stop, color):
        x_points = range(start, stop)
        y_points = [self.__calculate_y(x) for x in x_points]
        plt.scatter(x_points, y_points, marker='o', c=color)

        plt.show()


# instanciation
p1 = Polynome(4, 4, 4)
# tracé
p1.trace(-100, 100, "green")
```
